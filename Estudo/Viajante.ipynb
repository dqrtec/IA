```python
def distancia(a,b):
    x = abs(a[0]-b[0])
    y = abs(a[1]-b[1])
    return (x**2 + y**2)**(1/2)
```

```python
def somatorioDistancias(cidades):
    somatorio = 0
    for i in range( len(cidades) -1):
        somatorio += distancia(cidades[i], cidades[i+1])
    return somatorio
```

```python
def funcaoCusto(cidades):
    return somatorioDistancias(cidades)
        
```

```python
interacoes = 0
while interacoes != 100:
    interacoes += 1
    
    estadoInicial = cidades[:]
    estadoFinal = cidades[:]
    
    inicial = rdm.randint(1,9)
    final = rdm.randint(1,9)
    
    aux = estadoFinal[inicial]
    estadoFinal[inicial] = estadoFinal[final]
    estadoFinal[final] = aux
    
    if funcaoCusto(estadoFinal) < funcaoCusto(estadoInicial):
        cidades = estadoFinal
    
```
