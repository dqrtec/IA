```python
import random as rdn
import time
import datetime
```

```python
def temperaturaAtual(custoInicial, custoFinal):
    t = 10**((custoInicial - custoFinal)/temperatura)
    return t

def atualizarTemperatura():
    global temperatura
#     temperatura = temperatura-5
#     temperatura = temperatura**(1/2)
    if temperatura > 2:
        temperatura = temperatura/2
    else:
        temperatura = 1
```

```python
def criarTabelas(quantidadeTabelas):
    tabelas = [[] for i in range(quantidadeTabelas)]
    for i in range(quantidadeTabelas):
        tabelas[i] = criarTabela()
    return tabelas

def criarTabela():
    posicaoRainhas = [ j for j in range(n)]
    rdn.shuffle(posicaoRainhas)
    return posicaoRainhas

def copiarTabela(tabela):
    novaTabela = []
    for x in tabela:
        novaTabela.append(x)
    return novaTabela
```

```python
def custoLinhas(tabela):
    custoLinha = 0
    dicionario = {}
    for i in tabela:
        if i in dicionario:
            custoLinha += 1
        else:
            dicionario[i] = True
    return custoLinha

def custoDiagonalEsquerda(tabela, i):
    soma = 0
    superior = 0
    inferior = 0
    for j in range(0,i):
        if superior == 0:
            if tabela[i]+(j+1) < len(tabela):
                if tabela[i-(j+1)] == tabela[i]+(j+1):
                    superior += 1
        if inferior == 0:
            if tabela[i]-(j+1) >= 0:
                if tabela[i-(j+1)] == tabela[i]-(j+1):
                    inferior += 1
    return superior + inferior
                    
def custoDiagonalDireita(tabela, i):
    soma = 0
    superior = 0
    inferior = 0
    for j in range(i+1,len(tabela)):
        if superior == 0:
            if tabela[i]+(j-i) < len(tabela):
                if tabela[i+(j-i)] == tabela[i]+(j-i):
                    superior += 1
        if inferior == 0:
            if tabela[i]-(j-i) >= 0:
                if tabela[i+(j-i)] == tabela[i]-(j-i):
                    inferior += 1
    return superior + inferior

def custosDiagonais(tabela):
    somaDiagonal = 0
    for i in range(len(tabela)):
        somaDiagonal += custoDiagonalDireita(tabela,i)
        somaDiagonal += custoDiagonalEsquerda(tabela,i)
        
    return somaDiagonal/2

def custoTabela(tabela):
    custosLinhasTabela = custoLinhas(tabela)
    
    custosDiagonaisTabela = custosDiagonais(tabela)

    return custosLinhasTabela + custosDiagonaisTabela
```
