```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bmh")



N = 2
saudavel = (1342,0)
bacteria = (0, 2543)
viral    = (0, 1345)
ind = np.arange(N)    # the x locations for the groups
width = 0.85       # the width of the bars: can also be len(x) sequence
algStd = (0, 0)
importStd = (0, 0)
p1 = plt.bar(ind, saudavel, width)
p2 = plt.bar(ind, bacteria, width,
             bottom=saudavel)
p3 = plt.bar(ind, viral, width,
             bottom=bacteria)

plt.ylabel('Quantidade de EXAMES')
plt.title('Dados de TREINO')
plt.xticks(ind, ('Saudaveis', 'Pneumônicos'))
plt.yticks(np.arange(0, 5000, 500))
plt.legend((p1[0], p2[0], p3[0]), ('Saudavel','Bacteria', 'Virus'))

plt.show()
```
